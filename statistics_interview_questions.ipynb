### <u> Question 1</u> : Explain descriptive and inferential statistics

#### Descriptive statistics
* It deals with organizing and summarizing the data.
* It includes measures of central tendency - *Mean, median, mode* and measures of dispersion - *Range, variance, standard devaition, quantile deviation*

#### Inferential statistics

* It deals with collecting *sample* data, running experiements/tests on it for conclusions and inferences for the *population* data. 


### <u> Question 2</u> : Explain Population vs Sample

* Population is the entire dataset denoted by N. It is the entire group that you want to draw conclusions about
* Sample is the subset of the population that is actually used for the analysis,that gives inferences about the population. It is usually denoted by n.

### <u> Question 3</u> : Explain measures of central tendency

* Mean : Sum of observations / number of observations. It is the avergae of the observations
* Median : Central value in the sorted list of observations. A good way to tackle outliers.
    * Centre most observation in case of odd numbered list.
    * Average of two center elements in case of even numbered list.
* Mode : Most repeated value of an observation in a dataset. Element which has the maximum frequency.

### <u> Question 6</u> : Why does sample variance use (n-1) and not n , like population variance uses N?

* In statistics, **Bessel's correction** is the use of n − 1 instead of n in the formula for the sample variance and sample standard deviation, where n is the number of observations in a sample. 
* This method corrects the bias in the estimation of the population variance.

* Using n instead of (n-1) , there are chances that we underestimate the true population variance. Using (n-1) the error is reduced to an extent.

Note: 
> When you have an entire population and calculate any parameter (like the population variance or population standard deviation), your results will be accurate. That’s because you have all the data about your population. However, when you work with a sample, you’ve only got a small fraction of the population to work with. Therefore, your answers aren’t going to be as accurate as those you would have got, if you had the entire set of data to work with.

> In other words, sample calculations with n in the denominator are almost always going to be higher than calculations with n-1 in the denominator. When you subtract 1 from your sample size, it happens to turn out you’re making a fairly good adjustment for the deflated sum of squares figure as long as n isn’t huge
